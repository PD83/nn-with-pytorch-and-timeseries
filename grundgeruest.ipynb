```python
import pandas as pd
import torch
#from sklearn.datasets import load_boston
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable
import torch.optim as optim
import os
```

```python
class MeinNetz(nn.Module):
    def __init__(self):
        super(MeinNetz,self).__init__()
        self.lin1=nn.Linear(10,10)#(input,output)
        self.lin2=nn.Linear(10,10)
    
    def forward (self,x):
        x = F.relu(self.lin1(x)) # ab hier wird die variable lin1 als funktion behandelt mit übergabeparamater usw
        x = self.lin2(x)
        return x
    def num_flat_features(self,x):
        size = x.size()[1:]  # [1:],batch dimension. weil die erste spalte nur inizes sind glaub ich
        num = 1
        for i in size:
            num *= i  # anzahl der features, also zb 5zeilen mal 3 spalten
        return num
    
```

```python
class MeinNetz2(nn.Module):
    def __init__(self):
        super(MeinNetz2,self).__init__()
        self.lin1=nn.Linear(10,10)#(input,output)
        self.lin2=nn.Linear(10,10)
        self.lin2=nn.Linear(10,10)
    
    def forward (self,x):
        x = F.relu(self.lin1(x)) # ab hier wird die variable lin1 als funktion behandelt mit übergabeparamater usw
        x = self.lin2(x)
        return x
    def num_flat_features(self,x):
        size = x.size()[1:]  # [1:],batch dimension. weil die erste spalte nur inizes sind glaub ich
        num = 1
        for i in size:
            num *= i  # anzahl der features, also zb 5zeilen mal 3 spalten
        return num
```

```python
netz=MeinNetz()
#netz=netz.cuda()   # hierüber sagt man dass die gpu genutzt werden soll
#print(netz)
x =[1,0,0,0,1,0,0,0,1,1]
target = Variable(torch.Tensor([x for _ in range(10)])) 
input = Variable(target)
#input = input.cuda()
#target=target.cuda()
#input = Variable(torch.randn(10,10))  
#print (input)#tensor wird in variable verwandelt um sie in die funktionen geben zu können
             #torch.randn erstellt 10 mal 10 features
out = netz(input) # die klasse kann an dieser stelle als auch funktion genutzt werden,ist aber torchspezifisch
#print (out)
```

```python
x =[0,1,1,1,0,1,1,1,0,0]
#target ist das gewünschte ziel
target = Variable(torch.Tensor([x for _ in range(10)])) #hier wird erst ein tensor für x gebildet,und  dass dann als variable umgewandelt 
criterion = nn.MSELoss()  # mseloss ist eine fehlerfunktion,eine von vielen
loss = criterion(out,target) # das object criterion lässt sich hier auch wie ne funktion einsetzen # loss steht für fehler
print (loss)
#print (loss.grad_fn)
#print (loss.grad_fn.next_functions[0][0])  # so lässt sich der loss einens bestimmten layer ausgeben

netz.zero_grad() # setzt den fehlergradient auf null,sonst rechnet er den fehler von vorher mit?
loss.backward()  # hier wird das loss für das komplette netz erstellt?

optimizer= optim.SGD(netz.parameters(),lr=0.01)  # skochastic gradient descemt ,oder so. # parameters ist das was das training verursacht
#lr steht für learning rate. lässt sich dynamisch reduzieren. ist damit die schrittweite gemeint?

optimizer.step()
```
